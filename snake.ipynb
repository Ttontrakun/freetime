```python
import pygame
import random
import sys

# เริ่มต้นการตั้งค่าของ Pygame
pygame.init()

# กำหนดขนาดหน้าจอ
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Snake .io")

# สีที่ใช้ในเกม
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)
BLACK = (0, 0, 0)

# ตัวแปรสำหรับเกม
block_size = 10
snake_speed = 15
snake_body = [(100, 50), (90, 50), (80, 50)]  # หนอนเริ่มต้น
snake_direction = 'RIGHT'  # ทิศทางเริ่มต้น
food_position = (random.randrange(1, (screen_width // block_size)) * block_size,
                 random.randrange(1, (screen_height // block_size)) * block_size)

# ฟังก์ชันในการวาดหนอน
def draw_snake(snake_body):
    for segment in snake_body:
        pygame.draw.rect(screen, GREEN, pygame.Rect(segment[0], segment[1], block_size, block_size))

# ฟังก์ชันในการวาดอาหาร
def draw_food(food_position):
    pygame.draw.rect(screen, RED, pygame.Rect(food_position[0], food_position[1], block_size, block_size))

# ฟังก์ชันในการแสดงผลคะแนน
def display_score(score):
    font = pygame.font.SysFont(None, 35)
    score_text = font.render(f"Score: {score}", True, BLACK)
    screen.blit(score_text, (10, 10))

# เริ่มเกม
def game():
    global snake_body, snake_direction, food_position

    clock = pygame.time.Clock()
    score = 0
    game_over = False

    while not game_over:
        screen.fill(WHITE)

        # ตรวจสอบเหตุการณ์ต่าง ๆ ในเกม
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT and snake_direction != 'RIGHT':
                    snake_direction = 'LEFT'
                elif event.key == pygame.K_RIGHT and snake_direction != 'LEFT':
                    snake_direction = 'RIGHT'
                elif event.key == pygame.K_UP and snake_direction != 'DOWN':
                    snake_direction = 'UP'
                elif event.key == pygame.K_DOWN and snake_direction != 'UP':
                    snake_direction = 'DOWN'

        # เคลื่อนที่หนอน
        head_x, head_y = snake_body[0]
        if snake_direction == 'LEFT':
            head_x -= block_size
        elif snake_direction == 'RIGHT':
            head_x += block_size
        elif snake_direction == 'UP':
            head_y -= block_size
        elif snake_direction == 'DOWN':
            head_y += block_size

        new_head = (head_x, head_y)
        snake_body = [new_head] + snake_body[:-1]

        # ตรวจสอบการชนกับอาหาร
        if new_head == food_position:
            snake_body.append(snake_body[-1])  # ขยายหนอน
            food_position = (random.randrange(1, (screen_width // block_size)) * block_size,
                             random.randrange(1, (screen_height // block_size)) * block_size)
            score += 1

        # ตรวจสอบการชนกับตัวเองหรือขอบจอ
        if (new_head in snake_body[1:] or head_x < 0 or head_x >= screen_width or head_y < 0 or head_y >= screen_height):
            game_over = True

        # วาดหนอนและอาหาร
        draw_snake(snake_body)
        draw_food(food_position)
        display_score(score)

        pygame.display.flip()
        clock.tick(snake_speed)

    # แสดงคะแนนเมื่อเกมจบ
    font = pygame.font.SysFont(None, 50)
    game_over_text = font.render(f"Game Over! Final Score: {score}", True, BLACK)
    screen.blit(game_over_text, (screen_width // 4, screen_height // 3))
    pygame.display.flip()
    pygame.time.wait(2000)  # แสดงผลคะแนน 2 วินาที

    pygame.quit()
    sys.exit()

# เริ่มเกม
game()
```

```python
import pygame
import time
import random

# กำหนดขนาดของหน้าจอ
WIDTH = 640
HEIGHT = 480

# กำหนดสี
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (213, 50, 80)
GREEN = (0, 255, 0)
BLUE = (50, 203, 213)

# ขนาดของแต่ละบล็อก
BLOCK_SIZE = 10
SNAKE_SPEED = 20

# กำหนดความเร็วของเกม
clock = pygame.time.Clock()

# ฟังก์ชันเริ่มเกม
def gameLoop():
    game_over = False
    game_close = False

    # เริ่มต้นพิกัดของงู
    x1 = WIDTH / 2
    y1 = HEIGHT / 2

    x1_change = 0
    y1_change = 0

    snake_body = []
    snake_length = 1

    # สุ่มตำแหน่งของแอปเปิล
    foodx = round(random.randrange(0, WIDTH - BLOCK_SIZE) / 10.0) * 10.0
    foody = round(random.randrange(0, HEIGHT - BLOCK_SIZE) / 10.0) * 10.0

    # สร้างหน้าต่างของเกม
    pygame.init()
    gameDisplay = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption('Snake Game')

    font_style = pygame.font.SysFont("bahnschrift", 25)
    score_font = pygame.font.SysFont("comicsansms", 35)

    # ฟังก์ชันแสดงคะแนน
    def Your_score(score):
        value = score_font.render("Score: " + str(score), True, WHITE)
        gameDisplay.blit(value, [0, 0])

    # ฟังก์ชันแสดงข้อความเมื่อเกมจบ
    def message(msg, color):
        mesg = font_style.render(msg, True, color)
        gameDisplay.blit(mesg, [WIDTH / 6, HEIGHT / 3])

    # ลูปหลักของเกม
    while not game_over:

        while game_close:
            gameDisplay.fill(BLUE)
            message("You Lost! Press Q-Quit or C-Play Again", RED)
            Your_score(snake_length - 1)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        gameLoop()

        # ตรวจสอบการกดปุ่ม
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x1_change = -BLOCK_SIZE
                    y1_change = 0
                elif event.key == pygame.K_RIGHT:
                    x1_change = BLOCK_SIZE
                    y1_change = 0
                elif event.key == pygame.K_UP:
                    y1_change = -BLOCK_SIZE
                    x1_change = 0
                elif event.key == pygame.K_DOWN:
                    y1_change = BLOCK_SIZE
                    x1_change = 0

        # การตรวจสอบชนกำแพงและให้ "ทะลุ" ข้ามไปอีกด้าน
        if x1 >= WIDTH:
            x1 = 0
        elif x1 < 0:
            x1 = WIDTH - BLOCK_SIZE
        if y1 >= HEIGHT:
            y1 = 0
        elif y1 < 0:
            y1 = HEIGHT - BLOCK_SIZE

        x1 += x1_change
        y1 += y1_change
        gameDisplay.fill(BLUE)
        pygame.draw.rect(gameDisplay, GREEN, [foodx, foody, BLOCK_SIZE, BLOCK_SIZE])
        snake_head = []
        snake_head.append(x1)
        snake_head.append(y1)
        snake_body.append(snake_head)
        if len(snake_body) > snake_length:
            del snake_body[0]

        for block in snake_body[:-1]:
            if block == snake_head:
                game_close = True

        for segment in snake_body:
            pygame.draw.rect(gameDisplay, WHITE, [segment[0], segment[1], BLOCK_SIZE, BLOCK_SIZE])

        Your_score(snake_length - 1)

        pygame.display.update()

        # ตรวจสอบว่าแอปเปิลถูกกินหรือไม่
        if x1 == foodx and y1 == foody:
            foodx = round(random.randrange(0, WIDTH - BLOCK_SIZE) / 10.0) * 10.0
            foody = round(random.randrange(0, HEIGHT - BLOCK_SIZE) / 10.0) * 10.0
            snake_length += 1

        clock.tick(SNAKE_SPEED)

    pygame.quit()
    quit()

# เริ่มเกม
gameLoop()
```
