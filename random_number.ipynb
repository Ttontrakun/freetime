```python
import random

def number_guessing_game():
    # กำหนดช่วงตัวเลขที่สุ่ม
    lower_bound = 1
    upper_bound = 100
    secret_number = random.randint(lower_bound, upper_bound)  # สุ่มเลข

    print(f"ยินดีต้อนรับเข้าสู่เกม 'ทายตัวเลข'!")
    print(f"โปรแกรมได้เลือกตัวเลขระหว่าง {lower_bound} ถึง {upper_bound} ให้ทาย!")

    attempts = 0
    bot_attempts = 0
    user_won = False
    bot_won = False

    # บอทใช้กลยุทธ์ในการทายแบบ binary search
    bot_lower = lower_bound
    bot_upper = upper_bound

    # เก็บเลขที่ทายแล้วของผู้เล่น
    user_guesses = []
    user_feedback = []  # เก็บผลตอบรับของการทายแต่ละครั้ง (สูง/ต่ำ)

    while True:
        # ให้ผู้เล่นกรอกคำตอบ
        try:
            guess = int(input("กรุณาทายตัวเลข: "))
        except ValueError:
            print("กรุณากรอกตัวเลขที่ถูกต้อง!")
            continue
        
        # บันทึกเลขที่ทายแล้วของผู้เล่น
        user_guesses.append(guess)

        # ตรวจสอบการทายของผู้เล่น
        attempts += 1
        if guess < secret_number:
            user_feedback.append("ต่ำเกินไป!")
            # แสดงผลเลขที่ทายไปแล้วของผู้เล่นก่อนคำตอบ
            print(f"คุณทายไปแล้ว: {', '.join(map(str, user_guesses))}")
            print("ตัวเลข ที่ทายต่ำเกินไป!")
        elif guess > secret_number:
            user_feedback.append("สูงเกินไป!")
            # แสดงผลเลขที่ทายไปแล้วของผู้เล่นก่อนคำตอบ
            print(f"คุณทายไปแล้ว: {', '.join(map(str, user_guesses))}")
            print("ตัวเลข ที่ทายสูงเกินไป!")
        else:
            print(f"ยินดีด้วย! ทายถูกต้องแล้ว! ตัวเลขคือ {secret_number}.")
            print(f"คุณใช้เวลาทั้งหมด {attempts} ครั้งในการทายตัวเลข")
            user_won = True

        # ให้ bot ทายตัวเลขโดยใช้ binary search
        bot_guess = (bot_lower + bot_upper) // 2
        bot_attempts += 1
        print(f"Bot ทาย: {bot_guess}")

        # บอทปรับขอบเขตช่วงการทายตามคำตอบของผู้เล่น
        if bot_guess < secret_number:
            print("Bot ทายต่ำเกินไป!")
            bot_lower = bot_guess + 1  # บอทปรับขอบเขตช่วงต่ำขึ้น
        elif bot_guess > secret_number:
            print("Bot ทายสูงเกินไป!")
            bot_upper = bot_guess - 1  # บอทปรับขอบเขตช่วงสูงลง
        else:
            print(f"Bot ทายถูกต้อง! ตัวเลขคือ {secret_number}.")
            print(f"Bot ใช้เวลาทั้งหมด {bot_attempts} ครั้งในการทายตัวเลข")
            bot_won = True

        # ตรวจสอบว่าใครชนะ
        if user_won or bot_won:
            if user_won:
                print("คุณชนะ!")
            elif bot_won:
                print("Bot ชนะ!")
                
            break

# เริ่มเกม
number_guessing_game()
```
